```python
a = "purple is the best city in the forest"
b = "there is an art to getting your way and throwing bananas on to the street is not it"
c = "it is not often you find soggy bananas on the street"
d = "green should have smelled more tranquil but somehow it just tasted rotten"
e = "joyce enjoyed eating pancakes with ketchup"
f = "as the asteroid hurtled toward earth becky was upset her dentist appointment had been canceled"
g = "to get your way you must not bombard the road with yellow fruit" 


docs_1 = [a, b, c, d, e]
docs_2 = [d, e, f, g]
```

```python
from pprint import pprint

data_1 = []

for i in range(len(docs_1)):
    temp = {}
    temp["id"] = i+1
    temp["content"] = docs_1[i]
    data_1.append(temp)

print(data_1)

data_2 = []
for i in range(len(docs_2)):
    temp = {}
    temp["id"] = i+4
    temp["content"] = docs_2[i]
    data_2.append(temp)
    
print(data_2)
    
```

```python
import json
import os

output_path = "/home/peng_luh/__git/search_l3s/search_l3s_search_srv/datasets/test-dataset"

if not os.path.exists(output_path):
    os.makedirs(output_path)

output_file = os.path.join(output_path, "data.json")

with open(output_file, "a") as dataset:
    data = json.dumps(data_1, indent=4)
    dataset.write(data)
```
